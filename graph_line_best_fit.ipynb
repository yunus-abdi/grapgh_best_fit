```python
x, y = np.loadtxt('kenya-climate-data-1991-2016-temp-degress-celcius.txt', unpack=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
def line(x, a, b, c):
    return(a*np.cos(x*(2*3.142*x + b))+ c)
```

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(line, x, y)
```

```python
popt
```

```python
x_line = np.arange(min(x), max(x), 1)
```

```python
y_line = line(x_line, popt[0], popt[1], popt[2])
```

```python
plt.plot(x_line, y_line, '--', color='red')
plt.scatter(x, y)
```
